```python
import os
import sys
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy.io
```

```python
class PCA_face():
    def __init__(self, facefilepath ,output_dir ):
        self.output_dir = output_dir
        self.mat = scipy.io.loadmat(facefilepath)
        self.faces = self.mat['faces']
        self.dim = self.faces.shape[0]
        self.nfaces = self.faces.shape[1]
        print(f'{self.nfaces} images were loaded to Memmory ({self.dim} dimension)')
        
        
    def split_data(self,train_ratio):
        idx = list(range(self.nfaces))
        np.random.shuffle(idx)
        print(idx)
        ntrain = int(self.nfaces * train_ratio)
        self.train_data = self.faces[:,idx[:ntrain]]
        self.test_data = self.faces[:,idx[ntrain:]]
        return self.train_data , self.test_data
                    
    def show_samples_per_class(self):
        fig , axs = plt.subplots(nrows=4, ncols=10, figsize = (15,5))
        axs = axs.ravel()
        for i in range(40):
            axs[i].imshow(np.reshape(self.faces[:,i*10],(64,64),order ='F'),cmap = 'gray')
            axs[i].axis('off')
        plt.show()

    def compute_avg_face(self):
        self.avg_face = np.mean(self.train_data, axis=1)
        self.avg_face = np.reshape(self.avg_face , (-1, 1))
        return self.avg_face

    
    def show_avg_face(self):
        
        avg_face = self.compute_avg_face()
        avg_face = np.reshape(avg_face,(64,64),order ='F')
        plt.imshow(avg_face , cmap='gray')
        plt.axis('off')
        plt.title('Average Face')
        plt.show()
        
        
    def compute_normal_faces(self):
#         sub_face_vector = np.zeros(self.faces.shape, dtype=np.float32)
        self.normal_faces = self.train_data - self.avg_face
        print(self.normal_faces.dtype)
        return self.normal_faces
    
    def compute_cov(self):
        self.covariance_matrix = np.dot(self.normal_faces, self.normal_faces.T)
        self.covariance_matrix /= self.nfaces
        return self.covariance_matrix
    
    
    def find_eigen_faces(self) :
        eigenVectors , eigenValues , _ = np.linalg.svd(self.covariance_matrix, full_matrices = False )
        sort = eigenValues.argsort()[::-1]
        self.eigenValues = eigenValues[sort]
        self.eigenVectors = eigenVectors[:,sort]
        return self.eigenValues,eigenVectors
    
    def show_eigen_faces(self,nrows,ncols) :
        fig , axs = plt.subplots(nrows=nrows, ncols=ncols, figsize = (15,5))
        axs = axs.ravel()
        for i in range(nrows*ncols):
            axs[i].imshow(np.reshape(self.eigenVectors[:,i],(64,64),order ='F'),cmap = 'gray')
            axs[i].axis('off')
        plt.show()
        
        
    def show_curve_k(self):
        plt.plot(np.cumsum(self.eigenValues)/np.sum(self.eigenValues))
        plt.xlabel('Number of Component')
        plt.ylabel('Reconstruction Accuracy')
        plt.show()
        
        
    def find_K(self, percent):
        for i in range(0, len(self.eigenValues)):
            if ( ( np.sum(self.eigenValues[:i]) / np.sum(self.eigenValues) ) * 100 ) > percent :
                print(f'{i} Component is enough to reconstruct with {percent}% accuracy' )
                return i

        return None       
        
        
    def apply(self,n,k):
        redcueImage =np.dot(self.test_data[:,n],self.eigenVectors[:,:k])
        return redcueImage
    
    
    def reconstruct(self, reduced_img , k):
        reconstructed_image =reduced_img.dot(self.eigenVectors[:,:k].T)
        return reconstructed_image
    
    
    def apply_and_reconstruct(self, n,k):
        
        r = self.apply(n,k)
        rec = self.reconstruct(r,k)
        rec = np.reshape(rec, (64,64), order= 'F')
        img = np.reshape( self.test_data[:,n] , (64,64) , order= 'F')
        
        plt.subplot(1,2,1)
        plt.imshow(img , cmap = 'gray')
        plt.title('Original Image')
        plt.subplot(1,2,2)
        plt.imshow(rec,cmap = 'gray')
        plt.title('reconstructed Image')
        
    
        
        
```

```python
pca_object = PCA_face('faces.mat','out')
pca_object.split_data(0.7)
```
